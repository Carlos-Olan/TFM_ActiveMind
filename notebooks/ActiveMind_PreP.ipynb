# Carga de Datos.

```python
#Anlisis preeliminar

#Subir el archivo Excel
from google.colab import files
import io
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Solicita al usuario subir el archivo Excel
uploaded = files.upload()

#Se extrae su nombre
filename = list(uploaded.keys())[0]

# Leer el archivo Excel en un DataFrame de pandas
df = pd.read_excel(io.BytesIO(uploaded[filename]))

# Mostrar las primeras filas para verificar la correcta importación de los datos
print("Primeras filas del DataFrame:")
print(df.head())

```

# Transformación y Codificación de Variables

```python
# 1. Reformateo del Timestamp: Convertir a formato datetime y luego formatear a 'YYYY-MM-DD HH:MM:SS'
if 'Timestamp' in df.columns:
    df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
    df['Timestamp'] = df['Timestamp'].dt.strftime('%Y-%m-%d')
else:
    print("La columna 'Timestamp' no se encuentra en el DataFrame.")

# 2. Codificación del Género: "Masculino" -> 0, "Femenino" -> 1.
if 'Género:' in df.columns:
    genero_mapping = {"Masculino": 0, "Femenino": 1}
    df['Género:'] = df['Género:'].map(genero_mapping)
else:
    print("La columna 'Género:' no se encuentra en el DataFrame.")

# 3. Codificación de "¿Cuál es tu nivel de actividad física semanal?"
# Mapeo:
# "No hago ejercicio regularmente" -> 0
# "1-2 veces por semana"            -> 1
# "3-4 veces por semana"            -> 2
# "5 o más veces por semana"        -> 3
col_actividad = "¿Cuál es tu nivel de actividad física semanal?"
if col_actividad in df.columns:
    actividad_mapping = {
        "No hago ejercicio regularmente": 0,
        "1-2 veces por semana": 1,
        "3-4 veces por semana": 2,
        "5 o más veces por semana": 3
    }
    df[col_actividad] = df[col_actividad].map(actividad_mapping)
else:
    print(f"La columna '{col_actividad}' no se encuentra en el DataFrame.")

# 4. Codificación de "¿Qué importancia tiene para ti la salud mental en tu rutina de ejercicio?"
# Escala: "Poco importante" -> 0, "Algo importante" -> 2, "Muy importante" -> 4
col_importancia = "¿Qué importancia tiene para ti la salud mental en tu rutina de ejercicio?"
if col_importancia in df.columns:
    importancia_mapping = {
        "Poco importante": 0,
        "Algo importante": 2,
        "Muy importante": 4
    }
    df[col_importancia] = df[col_importancia].map(importancia_mapping)
else:
    print(f"La columna '{col_importancia}' no se encuentra en el DataFrame.")

# 5. Codificación de "¿Usas alguna dispositivo tecnológico para hacer ejercicio o mejorar tu bienestar?"
# Mapeo: "No" -> 0, "Sí" -> 1.
col_wearable = "¿Usas alguna dispositivo tecnológico para hacer ejercicio o mejorar tu bienestar?"
if col_wearable in df.columns:
    wearable_mapping = {"No": 0, "Sí": 1}
    df[col_wearable] = df[col_wearable].map(wearable_mapping)
else:
    print(f"La columna '{col_wearable}' no se encuentra en el DataFrame.")

# 6. Codificación de "¿Qué tan cómodo te sentirías si una app personalizara tu música y entrenamientos según tu estado de ánimo?"
# Escala: "Nada cómodo, no me interesa" -> 0, "No muy cómodo, prefiero elegir todo por mí mismo" -> 1,
# "Algo cómodo, pero me gustaría tener control manual" -> 2, "Muy cómodo, me gustaría probarlo" -> 4.
col_comodidad = "¿Qué tan cómodo te sentirías si una app personalizara tu música y entrenamientos según tu estado de ánimo?"
if col_comodidad in df.columns:
    comodo_mapping = {
        "Nada cómodo, no me interesa": 0,
        "No muy cómodo, prefiero elegir todo por mí mismo": 1,
        "Algo cómodo, pero me gustaría tener control manual": 2,
        "Muy cómodo, me gustaría probarlo": 4
    }
    df[col_comodidad] = df[col_comodidad].map(comodo_mapping)
else:
    print(f"La columna '{col_comodidad}' no se encuentra en el DataFrame.")

# 7. Codificación de "Si existiera una app que integre ejercicio, salud mental y personalización musical con IA, ¿estarías interesado/a en probarla?"
# Escala: "No, no me interesa" -> 0, "Tal vez, dependiendo de sus funciones" -> 2, "Sí, me interesa mucho" -> 4.
col_interes = "Si existiera una app que integre ejercicio, salud mental y personalización musical con IA, ¿estarías interesado/a en probarla?"
if col_interes in df.columns:
    interes_mapping = {
        "No, no me interesa": 0,
        "Tal vez, dependiendo de sus funciones": 2,
        "Sí, me interesa mucho": 4
    }
    df[col_interes] = df[col_interes].map(interes_mapping)
else:
    print(f"La columna '{col_interes}' no se encuentra en el DataFrame.")

```

#Normalizcación sistematica de las preguntas abiertas

```python
col_musica = 'En caso de haber seleccionado otra en la pregunta anterior:'

# Diccionario de normalización
music_map = {
    'reguetón': 'Reguetón',
    'regueton': 'Reguetón',
    'cumbia': 'Cumbia',
    'banda': 'Banda',
    'jazz': 'Jazz',
    'musica de sonidos de la naturaleza': 'Ambiental',
    'sonidos de la naturaleza': 'Ambiental',
    'según el tipo de actividad': 'Otro',
    'peliculas': 'Otro',
    'techno': 'Otros estilos modernos',
    'indie': 'Otros estilos modernos',
    'latina': 'Otros estilos modernos',
    'cantautor': 'Otros estilos modernos',
    'clasica': 'Clásica',
    'no': 'No escucha música',
    'sin musica': 'No escucha música',
    'corridos': 'Corridos',
    '70-80': 'Décadas pasadas',
    'podcasts': 'Podcasts',
    'natación': 'Natación',
    'natacion': 'Natación',
    'Natacion': 'Natación',
    'Natación': 'Natación',
    'Crossfit': 'CrossFit',
    'CrossFit': 'CrossFit',
    'Taichi': 'Taichí/Kung Fu',
    'TaiChi': 'Taichí/Kung Fu',
    'Taichi y Kung Fu': 'Taichí/Kung Fu',
    'Baile': 'Baile',
    'Baile, TaiChi, Kung Fu': 'Baile',
    'Diferentes modalidades de baile': 'Baile',
    'Escalar': 'Escalada',
    'Gimnasia de mantenimiento': 'Gimnasio/Fuerza',
    'Gimnasio': 'Gimnasio/Fuerza',
    'subir escaleras': 'Caminata/Actividad Ligera',
    'Caminar': 'Caminata/Actividad Ligera',
    'Arreglo de casa, el más agotador': 'Tareas domésticas',
    'No aplica': 'No aplica'
}

# Función de limpieza
def normalizar_musica(texto):
    if pd.isna(texto):
        return texto
    texto = texto.strip().lower()
    for key, value in music_map.items():
        if key in texto:
            return value
    return 'Otro'

# Aplicar función si la columna existe
if col_musica in df.columns:
    df[col_musica] = df[col_musica].apply(normalizar_musica)
else:
    print(f"La columna '{col_musica}' no se encuentra en el DataFrame.")

# Mostrar resumen
print("\nCategorías normalizadas en tipo de música (otro):")
print(df[col_musica].value_counts(dropna=False))
```
